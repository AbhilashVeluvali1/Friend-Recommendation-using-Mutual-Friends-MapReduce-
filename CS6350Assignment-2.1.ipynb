```
txtFile = sc.textFile("/FileStore/tables/Friends_Dataset.txt")
```

```
friendsList= txtFile.map(lambda x: x.split())
friendsList.collect()
```

```
userId = friendsList.map(lambda x: x[0])
userId.take(10)
```

```
friendList = friendsList.map(lambda x: [] if len(friendsList)==1 else list(map(lambda x: int(x), friendsList[1].split(','))))
```

```
friends = txtFile.map(lambda x: x.split("\t")).map(lambda y: (y[0], list(y[1].split(","))))
friends.collect()
```

```
pair = friends.flatMap(lambda x: [(x[0], z) for z in x[1]])
pair.collect()
```

```
tPair = pair.join(pair)
totalFriends = tPair.map(lambda x: x[1]).filter(lambda y: y[0] != y[1])
totalFriends.take(10)
```

```
mutualFriends = totalFriends.subtract(tPair)
mutualFriendsPair = mutualFriends.map(lambda x: (x, 1))
mutualFriendsPair.take(10)
```

```
#Method to sort the users based on mutual connections 
def recSort(recs):
    truncatedRecs = sorted(recs, key=lambda x: (-x[1], x[0]))[:10]
    topUserIds = [rec[0] for rec in truncatedRecs]
    return topUserIds
```

```
recommendation = mutualFriendsPair.reduceByKey(lambda l, m: l + m).map(lambda x: (x[0][0], (x[0][1], x[1]))).groupByKey().map(lambda y: (y[0], recSort(y[1]))) 
```

```

DF = recommendation.toDF(["UserId","Top10Recommendations"])
DF.show(10)
```
